```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# 1.1. Считать данные с помощью pandas  
# Вывести на экран первые 5 строк  
df = df = pd.read_csv('laptop_price.csv', sep = ',', encoding='latin-1')
df.head()
```

```python
# 1.2. Создать новый признак delta_renovated, который будет содержать разницу в годах между 
# годом реновацией дома и годом постройки дома 

# какжется работаем все таки с домами, а не с ноутами
df = pd.read_csv('kc_house_data.csv', sep = ',', encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
# 1.2. Создать новый признак delta_renovated, который будет содержать разницу в годах между 
# годом реновацией дома и годом постройки дома
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.head()
```

```python
# В новом столбце появились отрицательные значение, поправим это
def Delta_adjustment(x):
    if x < 0:
        return "Nan"
    else:
        return x

df['delta_renovated'] = df['delta_renovated'].apply(Delta_adjustment)
df.head()
```

```python
# Сразу скорректированное значение можем получить следующим образом
df['delta_renovated2'] = np.where(df['yr_renovated'] == 0, 'Nan', df['yr_renovated'] - df['yr_built'])
df.head()
```

```python
# 1.3. Если реновации дома не было, то в новом признаке поставьте 0

# Уже предвосхитила это в пердыдущем пункте, но вместо 0 указала Nan. 
# Считаю свой вариант более правильным, так как реновация могла быть в том же году, что и построен дом, 
# и 0 в дельте мог бы путать
```

```python
# 1.3. Создайте признаки года продажи, месяца продажи  
df.insert(2, 'date2', df['date'].apply(lambda x: x.replace('T000000', '')))
df.head()
```

```python
df.insert(3, 'year', df['date2'].apply(lambda x: x[:4]))
df.head()
```

```python
df.insert(4, 'month', df['date2'].apply(lambda x: x[4:6]))
df.head()
```

```python
df.insert(5, 'day', df['date2'].apply(lambda x: x[6:]))
df.head()
```

```python
# Есть еще временные ряды в pd
df.insert(2, 'datetime', pd.to_datetime(df['date']))
df.head()
```

```python
# Для примера добавим год используя временной ряд
df.insert(5, 'year2', df['datetime'].dt.year)
df.head()
```

```python
# 1.4. Удалите признаки date, zipcode, lat, long 
df = df.drop(columns = ['date','date2', 'zipcode', 'lat', 'long', 'year2'])
df.head()
```

```python
# 3.1 Найдите среднюю стоимость домов в зависимости от количества спален  
# Отсортируйте от меньшей стоимости к большей 
df.groupby('bedrooms').agg({'price': 'mean'}).round(0).sort_values(by = 'price')
```

```python
# 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома 
df.groupby('condition').agg({'price': ['min', 'mean', 'max']}).round(0)
```

```python
# 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида 
df.pivot_table(index='waterfront', columns = 'view',
               values='id', aggfunc='count', fill_value = '-', 
               margins = True, margins_name = 'Total')
```

```python
# 3.4 Каких домов в зависимости от этажности и количества спален больше?  
pd.crosstab(df.floors, df.bedrooms)

# Одноэтажных с 3 спальнями
```

```python
df.groupby(['floors', 'bedrooms'])['bedrooms'].count().unstack().fillna('-')
```

```python
df.pivot_table(index='floors',
               columns='bedrooms',
               aggfunc={'bedrooms':len},
               fill_value= '-')
```

```python
#  3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома 
df.pivot_table(index='grade',
               columns='condition',
               values = 'price',
               aggfunc= 'median',
               fill_value= '-', margins = True
               )
# проверила в excel - ok
```
